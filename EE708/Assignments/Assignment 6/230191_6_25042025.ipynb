```python
'''
Assignment 6
Programming Question 1
'''

import tensorflow as tf
import tensorflow_datasets as tfds
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Load tf_flowers dataset with 80:20 train-test split
(train_ds, test_ds), metadata = tfds.load('tf_flowers',
                                         split=['train[:80%]', 'train[80%:]'],
                                         as_supervised=True,
                                         with_info=True)

# Get number of classes
num_classes = metadata.features['label'].num_classes
print(f"Number of classes: {num_classes}")

# Normalize pixel values using Rescaling layer
normalization_layer = tf.keras.layers.Rescaling(1./255)

# Batch size and image size
batch_size = 32
img_height = 128
img_width = 128

# Function to preprocess images (resize and normalize)
def preprocess(image, label):
    image = tf.image.resize(image, [img_height, img_width])
    image = normalization_layer(image)
    return image, label

# Prepare datasets for training and testing
train_ds = train_ds.map(preprocess).cache().shuffle(1000).batch(batch_size).prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.map(preprocess).batch(batch_size).prefetch(tf.data.AUTOTUNE)

# Build CNN model with specified architecture
model_cnn = tf.keras.Sequential([
    # Input shape: (128, 128, 3)
    tf.keras.layers.Conv2D(32, kernel_size=3, strides=1, padding='valid', activation='relu',
                          input_shape=(img_height, img_width, 3), name='conv2d_3'),
    # Output shape: (126, 126, 32)

    tf.keras.layers.MaxPooling2D(pool_size=2, strides=2, name='max_pooling2d_3'),
    # Output shape: (63, 63, 32)

    tf.keras.layers.Conv2D(64, kernel_size=3, padding='valid', activation='relu', name='conv2d_4'),
    # Output shape: (61, 61, 64)

    tf.keras.layers.MaxPooling2D(pool_size=2, strides=2, name='max_pooling2d_4'),
    # Output shape: (30, 30, 64)

    tf.keras.layers.Conv2D(128, kernel_size=3, padding='valid', activation='relu', name='conv2d_5'),
    # Output shape: (28, 28, 128)

    tf.keras.layers.MaxPooling2D(pool_size=2, strides=2, name='max_pooling2d_5'),
    # Output shape: (14, 14, 128)

    tf.keras.layers.Flatten(name='flatten_1'),
    # Output shape: (25088)

    tf.keras.layers.Dense(128, activation='relu', name='dense_2'),
    # Output shape: (128)

    tf.keras.layers.Dropout(0.5, name='dropout_1'),
    # Output shape: (128)

    tf.keras.layers.Dense(num_classes, activation='softmax', name='dense_3')
    # Output shape: (num_classes)
])

# Compile the model
model_cnn.compile(optimizer='adam',
                 loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 metrics=['accuracy'])

# Print model summary
model_cnn.summary()

# Train the model
history = model_cnn.fit(train_ds,
                       validation_data=test_ds,
                       epochs=10)

# Evaluate the model
loss, accuracy = model_cnn.evaluate(test_ds)
print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")

# Plot training and testing accuracy over epochs
plt.figure(figsize=(10, 6))
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Test Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training and Testing Accuracy')
plt.legend()
plt.show()

# Generate classification report and confusion matrix
true_labels = []
pred_labels = []

for images, labels in test_ds:
    predictions = model_cnn.predict(images)
    pred_labels.extend(np.argmax(predictions, axis=1))
    true_labels.extend(labels.numpy())

print("Classification Report:")
print(classification_report(true_labels, pred_labels))

cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=metadata.features['label'].names,
           yticklabels=metadata.features['label'].names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
'''
Assignment 6
Programming Question 2
'''

import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Load stock price data
stock_data = pd.read_csv('A6.csv', header=0)
prices = stock_data.values.astype('float32')

# Normalize prices
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# Create sequences of 60 days input and 1 day output
sequence_length = 60
X, y = [], []
for i in range(sequence_length, len(prices_scaled)):
    X.append(prices_scaled[i-sequence_length:i, 0])
    y.append(prices_scaled[i, 0])
X, y = np.array(X), np.array(y)

# Reshape X for LSTM input: (samples, time_steps, features)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

# Split into train and test sets (80:20)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Build LSTM model with the specified architecture
model_lstm = tf.keras.Sequential([
    tf.keras.layers.LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], 1), name='lstm_2'),
    # Output shape: (None, 60, 50)

    tf.keras.layers.LSTM(50, name='lstm_3'),
    # Output shape: (None, 50)

    tf.keras.layers.Dense(25, name='dense_2'),
    # Output shape: (None, 25)

    tf.keras.layers.Dense(1, name='dense_3')
    # Output shape: (None, 1)
])

# Create an instance of the Adam optimizer
optimizer = tf.keras.optimizers.Adam()

# Compile the model using the optimizer instance
model_lstm.compile(optimizer=optimizer, loss='mean_squared_error')


# Train the model
history_lstm = model_lstm.fit(X_train, y_train,
                             epochs=20,
                             batch_size=32,
                             validation_data=(X_test, y_test))

# Plot training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(history_lstm.history['loss'], label='Train Loss')
plt.plot(history_lstm.history['val_loss'], label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Testing Loss')
plt.legend()
plt.show()


def display_full_model_summary(model):
    # Standard model summary
    model.summary()

    # Calculate model parameters
    trainable_count = np.sum([tf.keras.backend.count_params(w) for w in model.trainable_weights])
    non_trainable_count = np.sum([tf.keras.backend.count_params(w) for w in model.non_trainable_weights])
    total_count = trainable_count + non_trainable_count

    # Calculate memory usage
    trainable_bytes = trainable_count * 4  # float32
    non_trainable_bytes = non_trainable_count * 4
    total_bytes = trainable_bytes + non_trainable_bytes

    # Format bytes nicely
    def format_size(bytes):
        if bytes < 1024:
            return f"{bytes} B"
        elif bytes < 1024 * 1024:
            return f"{bytes / 1024:.2f} KB"
        else:
            return f"{bytes / (1024 * 1024):.2f} MB"

display_full_model_summary(model_lstm)

# Predict on test set
predictions = model_lstm.predict(X_test)

# Inverse transform predictions and true values to original scale
predictions_orig = scaler.inverse_transform(predictions)
y_test_orig = scaler.inverse_transform(y_test.reshape(-1, 1))

# Plot actual vs predicted stock prices
plt.figure(figsize=(12, 6))
plt.plot(y_test_orig, label='Actual Price')
plt.plot(predictions_orig, label='Predicted Price')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.title('Stock Price Prediction')
plt.legend()
plt.show()
```
